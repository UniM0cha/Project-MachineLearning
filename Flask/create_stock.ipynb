```python
import random
import pandas as pd
# 빈 DataFrame 생성하기
df = pd.DataFrame(columns=['stock_id', 'stock','product_id','store_id'])

stock_id = 1
for store_id in range(1, 55):
    for product_id in range(0,33):
        stock = random.randint(1000, 20000)
        # DataFrame에 특정 정보를 이용하여 data 채우기
        df = df.append(pd.DataFrame([[stock_id, stock, product_id, store_id]], columns=['stock_id', 'stock','product_id','store_id']), ignore_index=True)
        stock_id += 1
# df.set_index('stock_id', inplace=True)
df.to_csv('../data/stock.csv')
```
