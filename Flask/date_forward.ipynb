```python
import pandas as pd

before_data = pd.read_csv('../data/sales_modified.csv')
before_data
```

```python
import datetime
from dateutil.parser import parse

# print(before_data['ds'].iloc[-1])

format = '%Y-%m-%d'
last_date = parse(before_data['ds'].iloc[-1])
print('last_date :', last_date)

now = datetime.datetime.today()
print('now :', now)

sub_days = (now - last_date).days

after_ds = pd.to_datetime(before_data['ds']) + datetime.timedelta(days=sub_days)
before_data['ds'] = after_ds
print(before_data)

before_data.to_csv('../data/sales_modified2.csv')
```
